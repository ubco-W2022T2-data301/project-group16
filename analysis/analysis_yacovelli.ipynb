### Data Quality
What Numbeo consider as a "1" for data quality. How many missing values until data quality is considered a "0"?

```python
# Creating a new column with the total number of NaN values for a specific row
df['NaN_tot'] = df.iloc[:].isnull().sum(axis=1)
```

```python
plt.figure(figsize=(10,5))
his = sns.histplot(x=df['NaN_tot'], hue=df['data_quality'], multiple='stack', binwidth=1)
his.set(xlabel='NaN count',
       ylabel='Count',
       title='Histogram of Cities NaN Count',
       xticks = range(0,33,1))

plt.show()
```

As we can see all columns can only have a maximum of 3 missing values for it to be considered good quality by numbeo. 

### Where our data is located
Let's visualise where our data is from.

```python
countries = gpd.read_file(gpd.datasets.get_path("naturalearth_lowres"))
```

```python
plt.rcParams['figure.dpi'] = 300
#Set of all the countries in the dataset
countries_to_highlight = set(df["country"])

#Manually adding countries with name discrepancies with geopandas that have a "1"
countriesManualAddList = ["United States of America","Kosovo","Czechia","Bosnia and Herz.","Dem. Rep. Congo",
                          "Falkland Is.","Dem. Rep. Congo","Falkland Is.","Côte d'Ivoire","Solomon Is.","Cyprus",
                          "S. Sudan","Trinidad and Tobago"]
for country in countriesManualAddList:
    countries_to_highlight.add(country)

#find the countries to highlight that are not in the df
highlighted = countries[~countries['name'].isin(countries_to_highlight)]

# Create a plot with all countries in grey and the highlighted countries in red
ax = countries.plot(color='lightgrey', edgecolor='black', linewidth=0.15)
highlighted.plot(ax=ax, color='red', edgecolor='black', linewidth=0.15)
ax.set_axis_off()
ax.set_title("Countries NOT included in the \"Global Cost of Living\" dataset (red)", fontdict={'fontsize': 10})
plt.show()
```

### Where the "good quality" data is located
We will now see how many countries have cities that have data consideredas a "1" (fewer than 3 missing values) under the "data quality" column. 

```python
goodquality = set(df[df['data_quality'] == 1 ]["country"])
goodquality.add("United States of America")

#Manually adding countries with name discrepancies with geopandas that have a "1"
countriesManualAddList = ["United States of America","Kosovo","Czechia","Côte d'Ivoire","Solomon Is.","Cyprus",
                          "Trinidad and Tobago"]
for country in countriesManualAddList:
    goodquality.add(country)

highlighted = countries[countries['name'].isin(goodquality)]

# Create a plot with all countries in grey and the highlighted countries in red
ax = countries.plot(color='lightgrey', edgecolor='black', linewidth=0.15)
highlighted.plot(ax=ax, color='green', edgecolor='black', linewidth=0.15)
ax.set_axis_off()
ax.set_title("Countries that include at least one city with a \"1\" under the\n \"data quality\" column in the \"Global Cost of Living\" dataset (green)", fontdict={'fontsize': 10})
plt.show()
```

There appears to be sufficient global representation of at least one city in the green countries that contain "good quality" (fewer than 3 missing values).

## Creating necessities indexes (food, housing & transport) RQ1

In this section we will sum all of the columns for food, transport and shelter. From there, if we divide by Kelowna's respective category and multiply by 100 it will be possible to compare other cities relative to Kelowna to answer the question of Kelowna's affordability relative to other cities.  

For example, since Kelowna's value will be set to 100 any cities' index higher would be more expensive. So an index of 110 would be about 10% more than Kelowna.  

#### Food index

```python
col_food = [ "x3 (McMeal)", 'x9 (milk)','x10 (bread)', 'x11 (rice)', 'x12 (eggs)', 'x13 (cheese local)',
    'x14 (chicken fillets)', 'x15 (beef round)', 'x16 (apples)',
    'x17 (bananas)', 'x18 (oranges)', 'x19 (tomatoes)', 'x20 (potatoes)',
    'x21 (onions)', 'x22 (lettuce)', 'x23 (water 1.5L)']

df = yac.kelowna_index(df,col_food,'food')

df.head()
```

`Food index grouped by country`

```python
country_food_index = df.groupby('country')['food_index'].mean().sort_values(ascending=False)
country_food_index.head()
```

#### Housing index 

```python
col_housing=['x48 (apartment 1BR city center)',
    'x49 (apartment 1BR outside center)', 'x50 (apartment 3BR city center)',
    'x51 (apartment 3BR outside center)', 'x52 (price per sqm city center)',
    'x53 (price per sqm outside center)',
    'x36 (basic utilities)']
df = yac.kelowna_index(df,col_housing,'housing')
df.head()
```

`Housing index grouped by country`

```python
country_house_index = df.groupby('country')['housing_index'].mean().sort_values(ascending=False)
country_house_index.head()
```

#### Transport index 

```python
col_transport=['x28 (One-Way ticket transit)', 'x29 (monthly pass transit)',
    'x30 (taxi start)', 'x31 (taxi 1km)', 'x32 (Taxi 1hr wait)',
    'x33 (Gas 1L)', 
    "x34 (Volkswagen Golf)", "x35 (Toyota Corolla Sedan)"]

df = yac.kelowna_index(df,col_transport,'transport')

df.head()
```

`Transport index grouped by country`

```python
country_transport_index = df.groupby('country')['transport_index'].mean().sort_values(ascending=False)
country_transport_index.head()
```

#### Creating a sub dataframe: indexes_df
Contains city, respective country and their indexes.

```python
indexes_df= pd.DataFrame({
    'city': df['city'],
    'country': df['country'],
    'Food Index': df['food_index'],
    'Housing Index': df['housing_index'],
    'Transport Index': df['transport_index']
})
indexes_df[indexes_df['country']=='Canada'].sort_values(by='Food Index',ascending=False).head()
```

```python
melted_df = pd.melt(indexes_df, id_vars=['city', 'country'], value_vars=['Food Index', 'Housing Index', 'Transport Index'], var_name='index_type', value_name='index_value')
```

```python
g = sns.FacetGrid(melted_df, col='index_type', col_wrap=3, sharex=False)
g.map(sns.histplot,'index_value')

for ax in g.axes.flat:
    ax.tick_params(labelsize=15)

g.set(xlim=(0, 400))
g.set_titles(col_template="{col_name} (Kelowna red line)", size=20)
g.map(plt.axvline, x=100, color='red', linewidth=0.75)
g.fig.set_figwidth(20)
g.fig.set_figheight(12)
g.set_xlabels('Index Value', size=20)
g.set_ylabels('Count', size=20)
plt.show()
```

As these plots show, the only category where Kelowna comparably more affordable in is the transport index.
