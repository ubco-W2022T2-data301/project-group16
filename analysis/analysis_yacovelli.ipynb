### Comparing the price of a McDonald's meal to the cost of living index by COUNTRY

First we need to group the McDonald's meal by country and get the mean to compare these results to the coast of living index by country.

```python
mcMeal_index = ( df
                    .groupby('country')['x3 (McMeal)']
                    .mean()
                    .sort_values(ascending=False)
                    .to_frame()
                    .reset_index()
)

mcMeal_index = mcMeal_index.rename(columns = {
    'x3 (McMeal)': 'McMeal'})

mcMeal_index.head()
```

Since we want to analyze the cost of living index by **country**, we can create another data frame that has country, the three specidic indexes, the overall index along with the average cost of a McDonald's meal by county. 

```python
from functools import reduce

dfs = [country_transport_index, country_food_index, country_house_index, country_COL_index, mcMeal_index]
country_COL_df = reduce(lambda l,r: pd.merge(l,r,on='country'), dfs)

country_COL_df.head()
```

Now that we have our data that includes all the indixes and the cost of a combo meal at McDonald's grouped by country into one dataframe, let's visualize the correlation between each cost of living index and the McMeal to answer our question.

```python
melted_data = country_COL_df.melt(id_vars='McMeal', value_vars=['transport_index', 'food_index', 'housing_index', 'Weighted Cost of Living Index'])
plot = sns.FacetGrid(melted_data, col="variable", hue="variable", sharey=False, height=5)
plot.map(sns.regplot, "McMeal", "value", scatter_kws={'alpha': 0.3, 's':20}, x_jitter=0.1, y_jitter=0.1)
plot.fig.suptitle('Cost of Living indexes vs McMeal by Country', y=1.1, fontsize=22)

#customizing by grid
titles = ['Transport Index vs McMeal', 'Food Index vs McMeal', 'Housing Index vs McMeal', 'Weighted Cost of Living Index vs McMeal']
for ax, title in zip(plot.axes.flat, titles):
    ax.set_title(title,fontsize=18)
    ax.set_xlim(2.5,)
    ax.set_ylim(0,400)

#Remove redundant Y labels, keep leftmost 
for ax in plot.axes.flat[1:]:
    ax.set_yticklabels([])

#calculating correlation coefficient 
for ax, col in zip(plot.axes.flat, melted_data['variable'].unique()):
    #data for the current facet grid
    data = melted_data[melted_data['variable'] == col]
    #orrelation coefficient between 'McMeal' and 'value' from melted df
    corr = np.corrcoef(data['McMeal'], data['value'])[0][1]
    ax.text(0.95, 0.75, f'correlation={corr:0.2f}', ha='right', va='top', transform=ax.transAxes, fontsize=20,
            bbox=dict(facecolor='white', edgecolor='black', boxstyle='square,pad=0.2'))

plot.axes[0][0].set_ylabel('Index Value', fontsize=18)
plt.show()
```

As we can see there is a positive correlation between the indexes and the cost of a McDonald's meal indicating that the two move in the same direction. Beyond this fact, the Transport Index along with the Housing Index has a value of 0.27 indicating a weaker association between the two variables as compared to the other two indexes. Unsurprisingly, the Food Index vs McMeal has a stronger correlation of 0.74 meaning that the price of a McDonald's meal is a decent indicator of the cost of food in a country.

When analyzing the overall Weighted Cost of Living Index by country with the meal it has a correlation value of 0.47. This indicates a moderate positive relationship between these two variables and although correlation does not imply causation, the cost of living has many elements to it and one item as simple as a fast food combo cannot accurately depict the affordability of a country.

## RQ 3: How the cost of different types of transportation (public transit, taxi, personal car) affect a country's affordability & average monthly salary.
